## Подготовка данных.

### Заменим названия столбцов (приведём к нижнему регистру).

```python
df.columns = [x.lower() for x in df.columns]
```

### Преобразуем данные в нужные типы.

#### Столбец 'year_of_release'.
Преобразуем тип данных столбца year_of_release к целочисленному (int), т.к. год не может быть дробным (float).

```python
df['year_of_release'] = df['year_of_release'].astype('Int64')
```

#### Столбец 'user_score'.
Тип столбца 'User_Score' object, хотя оценка пользователей даётся в значениях типа float. Это связано с тем, что в данном столбце присутствует аббревиатура 'tbd' (to be defined - подлежащий определению), что означает, что оценки ещё не выставлены. Для удобства заменим 'tbd' на NaN и затем преобразуем тип данных.  

```python
df['user_score'] = df['user_score'].replace('tbd', np.NaN)
df['user_score'] = df['user_score'].astype('float')
```

### Обработаем пропуски.

#### Пропуски в столбце 'name'.
Выведем строки с пропусками.


```python
df[df['name'].isna()].head()
```

Количесвто пропусков очень мало - всего 2 штуки (или 0,012 %). От них можно избавиться, удалив данные строки. Кроме того, столбец 'genre' (как мы видили во время общего ознакомления с датафреймом) также имеет всего 2 пропуска и строки с пропусками у столбцов  'name' и 'genre' совападают. 

```python
df = df.dropna(subset = ['name'])
```

#### Пропуски в столбце 'year_of_release'.

Количесвто пропусков мало - 269 штук и составляет всего 1,6 %. От них можно избавиться, удалив данные строки.

```python
df = df.dropna(subset = ['year_of_release'])
```

#### Пропуски в столбце 'critic_score'.

В столбце с оценкой критиков слишком много пропусков. Их удалять нельзя. В данном случае эти пропсуки можно заменить на значение 'Unknown' (неизвестный). Данные пропсуки могут быть связаны с тем, что оценки данным играм не выставлялись. 

```python
df['critic_score'] = df['critic_score'].fillna('Unknown')
```

#### Пропуски в столбце 'user_score'.

Аналогично столбцу 'critic_score'.

```python
df['user_score'] = df['user_score'].fillna('Unknown')
```

#### Пропуски в столбце 'rating'.

Аналогично столбцам 'critic_score' и 'user_score'.

```python
df['rating'] = df['rating'].fillna('Unknown')
```

### Посчитаем суммарные продажи во всех регионах и запишим их в отдельный столбец.

```python
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

## Проверка гипотез.

Проверим следующие гипотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

Пороговое значение alpha примем равным 5%.

### Гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
Н0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Н1: Средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
df_XOne = df.query('year_of_release >= 2012').query('platform == "XOne"').copy()

df_XOne_user_score = df_XOne.query('user_score not in ["Unknown"]').copy()
df_XOne_user_score['user_score'] = df_XOne_user_score['user_score'].astype('float64')
df_XOne_user_score = df_XOne_user_score.query("platform == ['XOne']")['user_score']
```

```python
df_PC = games_2012_2016 = df.query('year_of_release >= 2012').query('platform == "PC"').copy()

df_PC_user_score = df_PC.query('user_score not in ["Unknown"]').copy()
df_PC_user_score['user_score'] = df_PC_user_score['user_score'].astype('float64')
df_PC_user_score = df_PC_user_score.query("platform == ['PC']")['user_score']
```

```python
alpha = 0.05  
results = st.ttest_ind(df_XOne_user_score, df_PC_user_score)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

<b> Вывод: </b>

Полученное значение p-value больше заданного нами порогового значение alpha, вследствии чего, мы не можем отвергнуть нулевую гипотезу - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

### Гипотеза - Средние пользовательские рейтинги жанров Action и Sports разные.
Н0: Средние пользовательские рейтинги жанров Action и Sports одинаковые.

Н1: Средние пользовательские рейтинги жанров Action и Sports разные.

```python
df_Action = df.query('year_of_release >= 2012').query('genre == "Action"')

df_Action_user_score = df_Action.query('user_score not in ["Unknown"]').copy()
df_Action_user_score['user_score'] = df_Action_user_score['user_score'].astype('float64')
df_Action_user_score = df_Action_user_score.query("genre == ['Action']")['user_score']
```

```python
df_Sports = df.query('year_of_release >= 2012').query('genre == "Sports"')

df_Sports_user_score = df_Sports.query('user_score not in ["Unknown"]').copy()
df_Sports_user_score['user_score'] = df_Sports_user_score['user_score'].astype('float64')
df_Sports_user_score = df_Sports_user_score.query("genre == ['Sports']")['user_score']
```

```python
alpha = 0.05  
results = st.ttest_ind(df_Action_user_score, df_Sports_user_score)
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

<b> Вывод: </b>
Полученное значение p-value меньше заданного нами порогового значение alpha, вследствии чего, мы отвергаем нулевую гипотезу и принимаем гипотезу -  Средние пользовательские рейтинги жанров Action и Sports разные.

Вторая Гипотеза - Средние пользовательские рейтинги жанров Action и Sports разные.

Н0: Средние пользовательские рейтинги жанров Action и Sports одинаковые.
